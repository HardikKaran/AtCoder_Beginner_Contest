# **Problem Statement**
You are given integers _N_ and _M_, a digit string _S_ of length _N_, and a digit string 
_T_ of length _M_. Here, a digit string is a string consisting of digits from 0 to 9.

You can perform the following operation zero or more times:

- Choose one character from _T_ and increase the chosen digit by 1. However, if the chosen digit is 9, change it to 0.

Find the minimum number of operations required to make _T_ a substring (contiguous subsequence) of _S_.

# **Constraints**
- 1 ≤ _M_ ≤ _N_ ≤100
- _N_ and _M_ are integers.
- _S_ is a digit string of length _N_.
- _T_ is a digit string of length _M_.

```python
import numpy as np

def operations(start, target):
    if target >= start:
        cost = target - start
    else:
        cost = (10 + target) - start
    
    return cost

def Substring2(N, M, S, T):
    costList = np.zeros((N - M) + 1)
    for i in range((N - M) + 1):
        for j in range(M):
            costList[i] += operations(int(T[j]), int(S[j + i]))

    return np.min(costList)

print(Substring2(4,2,"2025","91"))
print(Substring2(3,2,"438","38"))
print(Substring2(5,5,"00000","11111"))
print(Substring2(8,3,"20251227","438"))
```
